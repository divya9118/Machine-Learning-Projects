Under Sampling 

Build sample dataset containing similar distribution of normal and fraud transaction

No. of fraud 85
Take normal transaction : 85 and join them




```python
legit_sample=legit.sample(n=85)
```

```python
#concanate two dataset
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
#to check the difference
```

```python
X=new_dataset.drop(columns=['Class'],axis=1)
Y=new_dataset['Class']
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
X_train_pred=model.predict(X_train)
training_pred=accuracy_score(X_train_pred,Y_train)
print(training_pred)
```

```python
X_test_pred=model.predict(X_test)
test_pred=accuracy_score(X_test_pred,Y_test)
print(test_pred)
```
